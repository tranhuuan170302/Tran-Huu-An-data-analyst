```python
import pandas as pd
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31020936/GDPlist.csv", encoding='unicode_escape')
df.head()
```

```python
print('rows: ', df.shape[0])
print('columns: ', df.shape[1])
```

```python
df.info()
```

# việt hóa các tên cột

```python
df.rename(columns={'Country':"nuoc", 'Continent': 'chauluc', 'GDP (millions of US$)': 'GDP'}, inplace= True)
```

```python
df.head()
```

# chèn cột thanhpho vao sau cot nuoc

```python
df.insert(1, 'thanhpho', value= df['nuoc'])
```

```python
df.head()
```

# thay thế tên các nước thành thủ đô trong cột thanhpho

```python
df_2 = pd.read_csv("/country-list.csv")
df_2.head()
```

```python
for i in range(len(df['nuoc'])):
  for j in range(len(df_2['country'])):
    if str(df.iloc[i,0]).lower().strip() == str(df_2.iloc[j,0]).lower().strip():
      df['thanhpho'].replace(df.iloc[i,0], df_2.iloc[j,1], inplace = True)
    
```

```python
df.head()
```

# remove hangf co Asia

```python
df.drop(df[df.chauluc == 'Asia'].index, axis=0, inplace=True)
df.head()
```

# xóa bảng ghi có GDP trên 300000

```python
df.drop(df[df.GDP > 300000].index, axis = 0, inplace=True)
df.head(10)
```
