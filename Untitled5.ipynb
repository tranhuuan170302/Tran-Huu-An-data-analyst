```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

### Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12

```python
import datetime as dt
```

```python
df['InvoiceDate'].dt.month
```

```python
value = {
    1:1,
    2:1,
    3:1,
    4:2,
    5:2,
    6:2,
    7:3,
    8:3,
    9:3,
    10:4,
    11:4,
    12:4
}
```

```python
type(df['InvoiceDate'].dt.month[1])
```

```python


df['previous'] = df['InvoiceDate'].dt.month.map(value).fillna('Other')
```

```python
df[['InvoiceDate', 'previous']]
```

### Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice

```python
df['Amount'] = df['Quantity']* df['UnitPrice']
```

```python
df.head()
```

### Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.

```python
value_country = {'United Kingdom': "10%",
                 'France': "4.5%"
                 }
```

```python
df['Discount'] = df['Country'].map(value_country).fillna('0%')
```

```python
df[['Country', 'Discount']]
```

### Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.

```python
df['Discount1'] = df['Discount'].str.replace(r'(%)', '', regex=True).to_list()
```

```python
df['Discount1'] = df['Discount1'].astype('float')
```

```python
df['Amount'] = df['Amount'].astype("float")
```

```python
df['Total'] = df['Amount'] - (df['Discount1']/100)
```
