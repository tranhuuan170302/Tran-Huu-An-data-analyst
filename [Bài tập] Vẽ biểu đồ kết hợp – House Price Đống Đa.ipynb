# Vẽ biểu đồ xu hướng phân tích mối quan hệ giữa giá nhà với diện tích, giá nhà với số lượng phòng ngủ, nhận xét.

```python
df2 = df[['area', 'price', 'bedroom']]
```

```python
df2.info()
```

```python
df2.head()
```

```python
# x and y given as array_like objects
import plotly.express as px
fig = px.scatter(df2, x= "area", y= "price", trendline="ols", title = "graph relationship area by price")
fig.show()
```

```python
fig = px.scatter(df2, x= "bedroom", y= "price", trendline="ols", title = "graph relationship bedroom by price")
fig.show()
```

# Vẽ biểu đồ phân bố thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
df3 = df[["price", "balcony_direction"]]
```

```python
df3.info()
```

```python
df3 = df3.dropna()
```

```python
df3.info()
```

```python
df3 = df3.sort_values(by = "balcony_direction")
```

```python
fig = px.line(df3, x="balcony_direction", y="price", title='graph relationship price to balcony direction')
fig.show()
```
