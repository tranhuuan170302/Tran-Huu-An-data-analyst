# Vẽ các biểu đồ đường phân tích xu hướng giá gạo (Rice-Retail), giá Fuel (gas) trung bình cả nước (National Average) trong năm 2016, 2018, 2019 tại Thổ Nhĩ Kì.

```python
df_riceAndFuel = df[((df['ProductName'] == 'Rice - Retail')& ((df['Year'] == 2016) | (df['Year'] == 2018) | (df['Year'] == 2019) )) | ((df['ProductName'] == 'Fuel (gas) - Retail')& ((df['Year'] == 2016) | (df['Year'] == 2018) | (df['Year'] == 2019) ))  ]
```

```python
df_riceAndFuel.head()
```

```python
df_riceAndFuel.Year.unique()
```

```python
df_riceAndFuel.ProductName.unique()
```

```python
idx_rice = df_riceAndFuel[df_riceAndFuel['ProductName'] == 'Rice - Retail']
```

```python
idx_gas = df_riceAndFuel[df_riceAndFuel['ProductName'] == 'Fuel (gas) - Retail']
```

```python
df_rice = df.iloc[idx_rice.index,:]
```

```python
df_rice
```

```python
df_gas = df.iloc[idx_gas.index,:]
```

```python
df_gas
```

```python
lstgas = []
year = [2016,2018,2019]
for i in range(len(year)):
    dfy = df.iloc[df_rice[df_rice['Year'] == year[i]].index, :]
    dfy = dfy.astype({"Year":'str'})
    dfy = dfy.groupby('Month').sum()
    dfy = dfy.sort_values(by = 'Month')
    dfy = dfy.reset_index()
    lstgas.append(dfy)
```

```python
lstRice = []
year = [2016,2018,2019]
for i in range(len(year)):
    dfy = df.iloc[df_rice[df_rice['Year'] == year[i]].index, :]
    dfy = dfy.astype({"Year":'str'})
    dfy = dfy.groupby('Month').sum()
    dfy = dfy.sort_values(by = 'Month')
    dfy = dfy.reset_index()
    lstRice.append(dfy)
```

```python
lstgas[1]
```

```python
fig, ax = plt.subplots(2,figsize=(10,7))

ax[0].plot(lstRice[0]['Month'], lstRice[0]['Price'], '#575757')
ax[0].plot(lstRice[1]['Month'], lstRice[1]['Price'], '#FF0000')
ax[0].plot(lstRice[2]['Month'], lstRice[2]['Price'], '#008080')
ax[0].set_ylabel("Price", fontsize = 14)
ax[0].set_xlabel("month", fontsize = 14)
ax[0].set_title("price rice by month", fontsize = 20)
# legend
ax[0].legend(['2016', '2018', '2020'])

ax[1].plot(lstgas[0]['Month'], lstgas[0]['Price'], '#575757')
ax[1].plot(lstgas[1]['Month'], lstgas[1]['Price'], '#FF0000')
ax[1].plot(lstgas[2]['Month'], lstgas[2]['Price'], '#008080')
ax[1].set_ylabel("Price", fontsize = 14)
ax[1].set_xlabel("month", fontsize = 14)
ax[1].set_title("price gas by month", fontsize = 20)
# legend
ax[1].legend(['2016', '2018', '2020'])
# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```
