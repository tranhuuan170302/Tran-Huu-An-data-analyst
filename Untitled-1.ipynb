# Data preprocessing

```python
df1 = pd.get_dummies(df)
df1
```

```python
df1.info()
```

```python
truth_table = df1.isna().sum()
truth_table = truth_table.reset_index()
```

```python
columns_isNa = truth_table.loc[truth_table[0] >0, 'index']
columns_isNa
```

```python
columns_isNa = list(columns_isNa)
```

```python
columns_isNa
```

```python
def impute_median(series):
    return series.fillna(series.median())
```

```python
for i in range(len(columns_isNa)):
    df1[columns_isNa[i]] = df1[columns_isNa[i]].transform(impute_median)
```

# Handling Data

```python
y  = df1['SalePrice'].values
y
```

```python
#create feature array x
x = df1.drop("SalePrice", axis = 1).values
x
```

```python
x.shape
```

```python
y.shape
```

```python
#reshape y to have 1 column
y = y.reshape(-1,1)
y.shape
```

# Split Data

```python
# split the arrays into training and testing data sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

# model traing data

```python
# create a regressor object
LR = LinearRegression()
```

```python
# fit training set to the regerssion
LR.fit(x_train, y_train)
```

```python
# print model training linear regerssion 
print("Intercept = ", LR.intercept_)
print("coeficients = ", LR.coef_)
```

```python
# calculate R2-score
score = r2_score(y_test, y_pred)
print('r2-score = ', score)
print('mean_sqrd_error is  = ', mean_absolute_error(y_test, Y_prediction))
print('Root_mean_squared error of is =', np.sqrt(mean_absolute_error(y_test, Y_prediction)))
```
