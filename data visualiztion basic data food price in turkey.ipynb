# explain background

+ place: place of national "thổ nhĩ kỳ" and national average
+ productid : productid id in format int

+ UmId : Quantity of one product
+ UmName: product unit of measure 
+ month : that month of price 
+ year: that year of price
+ price: price of product in month/year

# Vẽ biểu đồ cột so sánh giá gạo (Rice-Retail) tháng 12 năm 2019 của Ankara, Istanbul, Izmir và National Average.

```python
df_rice = df[(df['ProductName'] == 'Rice - Retail') & (df["Month"] == 12) & (df["Year"] == 2019)]
```

```python
df_rice.info()
```

+ plot matplotlib barchart

```python
df_rice.head()
```

```python
plt.figure(figsize = (10, 5))
plt.bar(df_rice["Place"], df_rice["Price"])
plt.title("Price Rice-Retail of 4 place", fontsize = 20)
plt.xlabel("Price rice-retail", fontsize = 14)
plt.ylabel("place", fontsize = 14)
```

# Vẽ biểu đồ đường phân tích xu hướng giá gạo (Rice-Retail) trung bình cả nước (National Average) trong năm 2019 tại Thổ Nhĩ Kì.

```python
df_national_avg = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Rice - Retail') & (df["Year"] == 2019)]
```

```python
df_national_avg.info()
```

```python
df_national_avg = df_national_avg.sort_values('Price')
```

```python
df_national_avg
```

```python
plt.figure(figsize = (10, 5))
plt.plot(df_national_avg["Price"], df_national_avg['Month'])
plt.title("graph trend price rice-retail", fontsize = 20)
plt.xlabel("price rice",  fontsize = 14)
plt.ylabel("Month", fontsize = 14)
plt.show()
```

# Vẽ biểu đồ Scatter phân tích mối liên quan giữa giá gạo và giá gas trung bình quốc gia (National Average) tại Thổ Nhĩ Kì

```python
df_riceAndGas = df[((df['Place'] == "National Average") & (df["ProductName"] == 'Rice - Retail')) | ((df['Place'] == "National Average") & (df["ProductName"] == 'Garlic - Retail'))]
```

```python
df_riceAndGas['ProductName'].unique()
```

```python
df_riceAndGas.head(40)
```

```python
df_gas = df_riceAndGas[df_riceAndGas["ProductName"] == 'Garlic - Retail']["Price"]
```

```python
df_gas
```

```python
df_rice = df_riceAndGas[df_riceAndGas["ProductName"] == 'Rice - Retail']["Price"]
```

```python
df_rice
```

# plot scarrter price of rice-retail and gas

```python
plt.figure(figsize = (10, 5))
plt.title("Relationship between gas and rice", fontsize = 20, color = 'r')
plt.scatter(df_gas, df_rice )
plt.xlabel("price gas", fontsize = 14)
plt.ylabel("price rice", fontsize = 14)
```
