# xóa bảng ghi có productid trùng nhau và cập nhật lại chỉ số

```python
df=df.drop_duplicates(['ProductId'], keep = 'last').reset_index(drop= True)
df
```

#Tách file chứa thông tin sản phẩm

```python
df_pro = df.loc[:, ['ProductId','ProductName','UmId','UmName']]
df_pro
```

# Tách file chứa thông tin giá với số dòng từ bản ghi 10 đến 20

```python
df_pro1 = df.loc[10:20, ['ProductId','ProductName','UmId','UmName']]
df_pro1
```

```python
df1 = pd.merge(df_pro, df_pro1, on='ProductId')
df1
```
